# Glass Exercise

### Import Modules

```python
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
import keras

from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
```

### Step 1: Find out how many glass types and how many samples for each glass type

```python
col_names = ['id','ri','na','mg','al','si','k','ca','ba','fe','glass_type']

glass = pd.read_csv('glass.csv', names = col_names, index_col='id')

glass_types = 0

for i in glass.glass_type.unique():
    print('Samples of ' + str(i) + ' are ' + str(len(glass.loc[glass['glass_type']==i])))
    glass_types += 1
    
print('There are ' + str(glass_types) + ' glass types.')
```

### Step 2: Find out the correlation between 'al' and 'ri'

```python
x = glass['al']
y = glass['ri']

m,b = np.polyfit(x, y, 1)

print('Correlation is: ' + str(np.corrcoef(glass.al,glass.ri)[0][1]))

print('Slope of linear regression line is ' + str(m) + ' and the intercept is ' + str(b))
print(m, b)

plt.scatter(x, y)
plt.show()
```

### Step 3: Make prediction of 'ri' based on 'al'

```python
print('Predicted ri is: ' + str(((glass.al)*m) + b))
```

### Step 4: Create a new column of the predicted 'ri' values to the data frame

```python
glass['predicted_ri'] = ((glass.al)*m) + b
glass
```

### Step 5: Glass type 1, 2, 3 are window glass, glass type 5, 6, 7 are household glass. Cteate a new column called household that distinguishes household glass from window glass (window glass = 0, household glass = 1)

```python
glass['household'] = glass.glass_type.map({1:0,2:0,3:0,5:1,6:1,7:1})
glass
```

### Step 6: Make a scatter plot of 'al' and 'household'. Are they correlated? What is the linear regression fit?

```python
x = glass['al']
y = glass['household']

plt.scatter(x,y)
```

### Step 7: Make prediction of household type based on 'al' and add that to a new column called 'household_pred'

```python
model = Sequential()
model.add(Dense(32, input_shape=(1,)))
model.add(Activation('tanh'))
model.add(Dropout(0.25))
model.add(Dense(250))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(250))
model.add(Activation('sigmoid'))
model.add(Dense(1))
model.add(Activation('linear'))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x, y,
          epochs=100)
```

```python
prediction = model.predict(x)
prediction = np.round_(prediction)

actual = y

#print(prediction)
#print(actual)

fail_counter = 0

"""
#print("Count prediction based on a dataset ")
for i in range(len(prediction)):
    # print("Predicted count: " + str(prediction[i]) + ' - Actual count:' + str(actual[i]) + ' - Correct Prediction: ' + str(prediction[i] == actual[i]))
    if prediction[i] != actual[i]:
        fail_counter += 1
    # print()
"""

print("Failed predictions: " + str(fail_counter))

glass['household_predict'] = prediction
glass
```

### Logistic Regression

```python
### Useful if dependent variable is a classification/binary value
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
feature_col=['al']
x = glass[feature_col]
y = glass.household
logreg.fit(x,y)
glass['household_log_class'] = logreg.predict(x)
```

```python
plt.scatter(glass.al,glass.household)
plt.scatter(glass.al,glass.household_predict,color='green',label='linear')
plt.scatter(glass.al,glass.household_log_class,color='red',label='logistic')
plt.xlabel('al')
plt.ylabel('household')
plt.legend()
plt.show()
```

```python
glass
```

```python
glass['household_pred_prob']=logreg.predict_proba(x)[:,1]
```

```python
plt.scatter(glass.al,glass.household)
plt.plot(glass.al,glass.household_pred_prob,color='yellow',label='lin reg')
plt.scatter(glass.al,glass.household_predict,color='green',label='linear')
plt.scatter(glass.al,glass.household_log_class,color='red',label='logistic')
plt.scatter(glass.al,glass.household_pred_prob,color='purple',label='log_prob')
plt.xlabel('al')
plt.ylabel('household')
plt.legend()
plt.show()
```
