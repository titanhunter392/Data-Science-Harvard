### Load Data

```python
import os
import tflearn
import pandas as pd
```

```python
filename = os.path.join('./iris.csv') # path of data
iris = np.loadtxt(filename,delimiter =',', usecols = (0,1,2,3),skiprows=1)
type(iris)

df = pd.DataFrame.from_csv(filename)
df
```

```python
def class_to_numeric(x):
    if x=='Iris-setosa':
        return 1
    if x=='Iris-versicolor':
        return 2
    if x=='Iris-virginica':
        return 3

df['class_num'] = df['class'].apply(class_to_numeric)
df

df.to_csv('out.csv')
```

```python
from tflearn.data_utils import load_csv
data, labels = load_csv('out.csv', target_column = 4, columns_to_ignore = [5], categorical_labels = False, n_classes = 1)
```

```python
net = tflearn.input_data(shape=[None,4])
net = tflearn.fully_connected(net, 32)
net = tflearn.fully_connected(net, 32)
net = tflearn.fully_connected(net, 32)
net = tflearn.fully_connected(net, 32)
net = tflearn.fully_connected(net, 1, activation='sigmoid')
net = tflearn.regression(net)

model = tflearn.DNN(net)
model.fit(data, labels, n_epoch=100, batch_size=30, show_metric=True)
```

### Step 1: Get Dimension of iris.csv

```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

filename = os.path.join('./iris.csv') # path of data
df = pd.read_csv(filename)
df
```

### Step 2: Find Min, Max, Average, and StD for each column

```python
sepal_length_min = min(df['sepal_length'])
sepal_length_max = max(df['sepal_length'])
sepal_length_average = (sum(df['sepal_length']))/((df['sepal_length']).count())
sepal_length_std = np.std(df['sepal_length'])

print('sepal length')
print(sepal_length_min)
print(sepal_length_max)
print(sepal_length_average)
print(sepal_length_std)

sepal_width_min = min(df['sepal_width'])
sepal_width_max = max(df['sepal_width'])
sepal_width_average = (sum(df['sepal_width']))/((df['sepal_width']).count())
sepal_width_std = np.std(df['sepal_width'])

print('sepal width')
print(sepal_width_min)
print(sepal_width_max)
print(sepal_width_average)
print(sepal_width_std)

pedal_length_min = min(df['pedal_length'])
pedal_length_max = max(df['pedal_length'])
pedal_length_average = (sum(df['pedal_length']))/((df['pedal_length']).count())
pedal_length_std = np.std(df['pedal_length'])

print('pedal length')
print(pedal_length_min)
print(pedal_length_max)
print(pedal_length_average)
print(pedal_length_std)

pedal_width_min = min(df['pedal_width'])
pedal_width_max = max(df['pedal_width'])
pedal_width_average = (sum(df['pedal_width']))/((df['pedal_width']).count())
pedal_width_std = np.std(df['pedal_width'])

print('pedal width')
print(pedal_width_min)
print(pedal_width_max)
print(pedal_width_average)
print(pedal_width_std)
```

### Step 3: Sort Datasets by 4th column

```python
df_4 = df.sort_values('pedal_width')
df_4
```

### Step 4: Give names to each columns:
'sepal_length','sepal_width','petal_length','petal_width','species'

```python
df.columns = ['sepal_length','sepal_width','petal_length','petal_width','species']
df
```

### Step 5: Create Histogram of 'petal_width' grouped by 'species'

```python
df1 = df[df['species'] == 'Iris-setosa']
df2 = df[df['species'] == 'Iris-versicolor']
df3 = df[df['species'] == 'Iris-virginica']

plt.hist(df1['petal_width'])
plt.show()

plt.hist(df2['petal_width'])
plt.show()

plt.hist(df3['petal_width'])
plt.show()
```

### Step 6: Define a new feature that represents the petal area

```python
area = df['petal_length'] * df['petal_width']
df['petal_area'] = area
df
```

### Bonus: Define a function that accepts a row of data and returns a predicted species. Make prediction of species based on petal area

```python
import keras

from keras.models import Sequential
from keras.layers import Dense, Activation, Dropout
```

```python
x_train = df.drop(["species"])
y_train = df[["species"]]

model = Sequential()
model.add(Dense(32, input_shape=(None, 101)))
model.add(Activation('relu'))
model.add(Dropout(0.25))
model.add(Dense(250))
model.add(Activation('relu'))
model.add(Dense(1))
model.add(Activation('linear'))
model.compile(optimizer='adam', loss='mse')
```

```python
model.fit(x_train, y_train,
          epochs=20,
          batchsize=128)
```
