```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sb

filename = os.path.join('./iris_data.csv') # path of data
iris_data = pd.read_csv(filename,na_values='na')
```

```python
sb.pairplot(iris_data.dropna(),hue='class')
plt.show()
```

```python
iris_data.loc[iris_data['class']=='versicolor','class']='Iris-versicolor'
iris_data.loc[iris_data['class']=='Iris-setossa','class']='Iris-setosa'
```

```python
iris_data['class'].unique()
```

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull())|
             (iris_data['sepal_width_cm'].isnull())|
             (iris_data['petal_length_cm'].isnull())|
             (iris_data['petal_width_cm'].isnull())]
```

```python
iris_data.loc[iris_data['class']=='Iris-setosa','petal_width_cm'].hist()
plt.show()
```

```python
average_petal_width = iris_data.loc[iris_data['class'] == 'Iris-setosa','petal_width_cm'].mean()
iris_data.loc[(iris_data['class']=='Iris_setosa')&
             (iris_data['petal_width_cm'].isnull()),
             'petal_width_cm'] = average_petal_width
```

```python
iris_data.loc[(iris_data['class']=='Iris_setosa')&
              (iris_data['petal_width_cm']==average_petal_width)]
```

```python
iris_data.loc[(iris_data['sepal_length_cm'].isnull())|
             (iris_data['sepal_width_cm'].isnull())|
             (iris_data['petal_length_cm'].isnull())|
             (iris_data['petal_width_cm'].isnull())]
```

```python
iris_data.dropna(inplace=True)
```

```python
iris_data.to_csv('iris_data_clean.csv',index=False)
```

```python
iris_clean = pd.read_csv('iris_data_clean.csv')
```

```python
sb.pairplot(iris_clean,hue='class')
plt.show()
```

```python
iris_clean.loc[(iris_clean['class']=='Iris-versicolor') &
              (iris_clean['sepal_length_cm'] < 1.0)]
```

```python
iris_clean.loc[(iris_clean['class']=='Iris-versicolor')&
              (iris_clean['sepal_length_cm']<1.0),
              'sepal_length_cm'] *= 100.0
```

```python
iris_clean.loc[iris_clean['class']=='Iris-versicolor','sepal_length_cm'].hist()
plt.show()
```

```python
filename = os.path.join('./euro2012.csv')
df = pd.read_csv(filename)
df
```

### Step 2: Create a new Data Frame showing 'Team', 'YellowCards', 'RedCards'

```python
col_list = ['Team','Yellow Cards','Red Cards']
newdf = df[col_list]
newdf
```

### Step 3: Sort the teams by 'RedCards', then to 'YellowCards'

```python
newdf = newdf.sort_values(['Red Cards','Yellow Cards'])
newdf
```

### Step 4: Filter teams that scored more than 6 goals

```python
df.loc[(df['Goals']) > 6.0]
```

# Step 5: Select teams that start with G

```python
df.loc[((df['Team']) >= 'G') & ((df['Team']) < 'H')]
```

### Step 6: Select teams that end with E

```python
teame = []

for i in df['Team']:
    temp = i[::-1]
    if temp[0] == 'e':
        teame.append(i)

teame
```
